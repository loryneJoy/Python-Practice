# HYPOTHESIS TESTING
## EXPERIMENTAL DESIGN
## z-score and p-value

```python
from scipy import stats
import numpy as np
```

```python
#calculating probability up to z-score 1.5
print(stats.norm.cdf(1.5))
```

```python
# Probabilities greater than z-score of >= 1.5
print(1-stats.norm.cdf(1.5)) #p-value calculation
```

```python
# One sample z test
iq = 103
mu = 100
sigma = 16 
n = 40

z = (iq - mu)/(sigma/np.sqrt(n))
print (z)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('seaborn-v0_8-whitegrid')
plt.fill_between(x=np.arange(-4,1.19,0.01),
                 y1= stats.norm.pdf(np.arange(-4,1.19,0.01)) ,
                 facecolor='red',
                 alpha=0.35,
                 label= 'Area below z-statistic'
                 )

plt.fill_between(x=np.arange(1.19,4,0.01), 
                 y1= stats.norm.pdf(np.arange(1.19,4,0.01)) ,
                 facecolor='blue',
                 alpha=0.35, 
                 label= 'Area above z-statistic')
plt.legend()
plt.title ('z-statistic = 1.19')
plt.show()
```

```python
stats.norm.cdf(z)
```

```python
pvalue = 1 - stats.norm.cdf(z)
print (pvalue)
```

```python
x_bar = 75
n2 = 36
sd = 30
mu2 = 60

z_stat = (x_bar - mu2)/(sd/np.sqrt(n2))
print (z_stat)
```

```python

# pvalue
p_value2 = 1 - stats.norm.cdf(z_stat)
print(p_value2)
```

```python
scores = [434, 694, 457, 534, 720, 400, 484, 478, 610, 641, 
          425, 636, 454, 514, 563, 370, 499, 640, 501, 625, 
          612, 471, 598, 509, 531]


x_barstudents = sum (scores)/ len(scores) #sample mean
n_pop = 25 # no of students
sigma_sd = 100
pop_mean = 500

z_value_stats = (x_barstudents - pop_mean)/(sigma_sd/np.sqrt(n_pop))
print (z_value_stats)
```

```python
#p value

p_value_stats = 1 - stats.norm.cdf(z_value_stats)
print (p_value_stats)
```
