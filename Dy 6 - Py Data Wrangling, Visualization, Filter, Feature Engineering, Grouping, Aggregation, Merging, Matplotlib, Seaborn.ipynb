#### DATA WRANGLING & VISUALIZATION
# FILTERING
- filtering rows based on a condition
- We use boolean indexing to filter rows in python

```python
import pandas as pd #loads & reads files

import numpy as np #performs calculations

import matplotlib.pyplot as plt #for general visualizations eg pie chart

import seaborn as sns #for statistical visualization eg histogram
```

```python
df = pd.read_csv(r"titanic _dataset.csv")
df.head() #shows table 
```

```python
df.shape #gives number of rows and columns
```

```python
df.columns #shows columns
```

```python
df.info()
```

```python
df.describe()#gives statistics of the data
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median) #fillna fills missing values. here we fill with the median
df.info()
```

```python
#to convert age from object to numeric
df["Age"] = pd.to_numeric(df["Age"],errors="coerce")
df.info()
```

```python
#to get adults older than 18
adults = df[df["Age"]>= 18]
adults.head()
```

```python
#to show passengers in pclass 3
passengers = df[df["Pclass"] == 3]
passengers.head()
```

```python
passengers["Pclass"].value_counts() #filters only passengers in class 3
```

#### SORT VALUES - sorts data

```python
#to sort fare in ascending order 
df_sorted = df.sort_values(by="Fare", ascending=True)
df_sorted.head(10)
```

#### FEATURE ENGINEERING
- it's creating new columns from existing columns to improve analysis

```python
#to create age groups using pandas:
df["Age_group"] = pd.cut(df["Age"], bins = [0, 12, 19, 59, 100], labels= ["Child", "Teen", "Adult", "Senior"])
df.head()
```

```python
#To extract title from Name column:
df["Title"] = df["Name"].apply(lambda name: name.split(",")[1].split(".")[0].strip())
df.head()
```

#### GROUPING & AGGREGATION
- .Groupby () and .agg() to compute statistics for sub-groups
- in .Groupby, one has to be an object, the other numeric

```python
#Subgroup to show how many people survived per passenger class:
grouped = df.groupby("Pclass")["Survived"].agg(["count", "mean"]) #pclass is the object column, survived is the numeric column
grouped.head()
```

```python
#who survived and did not survive:
df["Survived"].value_counts()
```

```python
#table of the survivors:
df_survived = df[df["Survived"] == 1]
df_survived.head()
```

```python
#shows who survived per class:
df_survived["Pclass"].value_counts()
```

#### MERGING DATAFRAMES
- It's combining 2 tables. Is similar to joins in SQL

```python
#converting the data in Pandas dictionaries to dataframe:
df1 = pd.DataFrame({
    "ID": [1, 2, 3, 4, 5],
    "Name": ["Matilda", "Hillary", "Loryne", "Isaac", "Makena"]
})

df2 = pd.DataFrame({
    "ID": [1, 2, 3, 4, 5],
    "Age": [27, 29, 26, 27, 19]
})
```

```python
#to merge 2 tables based on common columns::
merged_df = pd.merge(df1, df2, on="ID")
merged_df.head()
```

```python
#An age histogram:
plt.hist(df["Age"], bins=20)
plt.title("Age Histogram")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```
